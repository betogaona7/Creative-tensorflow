```
from google_images_download import google_images_download as gid

response = gid.googleimagesdownload()
urls = response.download({"keywords":'churros', "limit":100, "format":'jpg', "output_directory":'./dataset/churros'})
```

```
from PIL import Image 

def resize(width, height, images, out_dir):
  for i in range(len(images)):
    img = Image.open(images[i])
    img = img.resize((width, height), Image.BILINEAR)
    img.save(out_dir + str(i) + '.jpg')

resize(250, 250, urls['churros'], './dataset/db_churros/')
```

```
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
import glob

data_path = './dataset/db_churros/*.jpg'
img_paths = glob.glob(data_path)

images = []
for img in glob.glob(data_path):
  images.append(plt.imread(img))
```

```
plt.imshow(images[0])
```

```
data = np.array(images)
data.shape
```

### Reduce down the entire dataset to a single representation which describes what most of our dataset looks likes.

```
mean_img = np.mean(data, axis=0)
plt.imshow(mean_img.astype(np.uint8))
```

### Display where and how much variance there is in our previous mean image representation. (where changes are likely to be in our dataset of images)

```
std_img =np.std(data, axis=0)
plt.imshow(std_img.astype(np.uint8))
```

```
plt.imshow(np.mean(std_img, axis=2).astype(np.uint8))
```
